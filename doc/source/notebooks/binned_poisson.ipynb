# Using Binned Poisson Likelihood

```python
import GPflow
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

## One dimensional example:  The coal mining disasters dataset in the UK from 1851 to 1962
R.G. Jarrett. A note on the intervals between coal mining disasters. Biometrika, 66:191â€“193, 1979.

```python
def load_coal():
    X = np.loadtxt('data/coal.csv')
    return X

X = load_coal()
year_bins = np.arange(np.floor(np.min(X)),np.ceil(np.max(X)))
Y, bb = np.histogram(X,year_bins)
Y = np.asarray(Y,dtype=np.double)
binsize = np.diff(year_bins.flat)[0] # Binsize is a year
```

```python
#build the model
k = GPflow.kernels.Matern32(1,ARD=False,active_dims=[0]) + GPflow.kernels.Bias(1)
l = GPflow.likelihoods.Poisson(binsize=binsize)
m = GPflow.gpmc.GPMC(year_bins[:-1,None], Y[:,None], k, l)
```

```python
m.kern.matern32.lengthscales.prior = GPflow.priors.Gamma(1., 1.)
m.kern.matern32.variance.prior = GPflow.priors.Gamma(1.,1.)
m.kern.bias.variance.prior = GPflow.priors.Gamma(1.,1.)
```

## Let's look at the data aggregated to the year

```python
plt.figure(figsize=(8,5))
plt.plot(year_bins[:-1],Y,'r.')
plt.title('Number of Coal Disasters per Year')
plt.xlabel('Year')
plt.ylabel('Number of Disasters')
plt.show()
```

```python
m.optimize(maxiter=15) # start near MAP
samples = m.sample(500, verbose=True, epsilon=0.12, Lmax=15)
```

```python
xtest = np.linspace(np.floor(np.min(X)),np.ceil(np.max(X)),100)[:,None]
f_samples = []
for s in samples:
    m.set_state(s)
    f_samples.append(m.predict_f_samples(xtest, 5))
f_samples = np.vstack(f_samples)
```

```python
rate_samples = np.exp(f_samples[:,:,0])
plt.figure(figsize=(12, 6))
line, = plt.plot(xtest, np.mean(rate_samples, 0), lw=2)
plt.fill_between(xtest[:,0], np.percentile(rate_samples, 5, axis=0), np.percentile(rate_samples, 95, axis=0), color=line.get_color(), alpha = 0.2)
plt.plot(year_bins[:-1], Y/binsize, 'kx', mew=2)
plt.ylim(-1.1, np.max(np.percentile(rate_samples, 95, axis=0)))
plt.plot(X,-1*np.ones_like(X),'b|')
plt.title('Coal Data with Binsize = 1 year')
plt.xlabel('Year')
plt.ylabel('Number of Disasters')
```

## Now let's use a larger binsize

```python
X = load_coal()
bins = np.linspace(np.min(X),np.max(X),50)
Y, bb = np.histogram(X,bins)
Y = np.asarray(Y,dtype=np.double)
binsize = np.diff(bins.flat)[0] # about 2.25 years
```

```python
#build the model
k = GPflow.kernels.Matern32(1,ARD=False,active_dims=[0]) + GPflow.kernels.Bias(1)
l = GPflow.likelihoods.Poisson(binsize=binsize)
m = GPflow.gpmc.GPMC(bins[:-1,None], Y[:,None], k, l)
```

```python
m.kern.matern32.lengthscales.prior = GPflow.priors.Gamma(1., 1.)
m.kern.matern32.variance.prior = GPflow.priors.Gamma(1.,1.)
m.kern.bias.variance.prior = GPflow.priors.Gamma(1.,1.)
```

```python
plt.figure(figsize=(8,5))
plt.plot(bins[:-1],Y,'r.')
plt.title('Number of Coal Disasters per Year')
plt.xlabel('Year')
plt.ylabel('Number of Disasters')
plt.show()
```

```python
m.optimize(maxiter=20) # start near MAP
samples = m.sample(500, verbose=True, epsilon=0.12, Lmax=15)
```

```python
xtest = np.linspace(np.floor(np.min(X)),np.ceil(np.max(X)),50)[:,None]
f_samples = []
for s in samples:
    m.set_state(s)
    f_samples.append(m.predict_f_samples(xtest, 5))
f_samples = np.vstack(f_samples)
```

```python
rate_samples = np.exp(f_samples[:,:,0])
plt.figure(figsize=(12, 6))
line, = plt.plot(xtest, np.mean(rate_samples, 0), lw=2)
plt.fill_between(xtest[:,0], np.percentile(rate_samples, 5, axis=0), np.percentile(rate_samples, 95, axis=0), color=line.get_color(), alpha = 0.2)
plt.plot(bins[:-1], Y/(binsize), 'kx', mew=2) # NOTE: We divide the counts by binsize!
plt.ylim(-1.1, np.max(np.percentile(rate_samples, 95, axis=0)))
plt.plot(X,-1*np.ones_like(X),'b|')
plt.title('Coal Data with Binsize = ~2.25 years')
plt.xlabel('Year')
plt.ylabel('Number of Disasters')
```

## Two dimensional example:  The pines dataset ??


```python
from itertools import product
def load_pines():
    return np.loadtxt('data/pines.csv', delimiter=',')

def getCounts(gridResolution):
    locations = load_pines()
    counts, _ = np.histogramdd(locations, bins=(gridResolution, gridResolution), range=((0, 1.), (0., 1.)))
    return counts.reshape(-1, 1)

def getGrid(gridResolution):
    linearValues = np.linspace(0., 1., gridResolution+1)
    binEdges = np.array([np.array(elem) for elem in product(linearValues, linearValues)])
    offsetValues = linearValues[:-1] + 0.5*np.diff(linearValues)[0]
    binMids = np.array([np.array(elem) for elem in product(offsetValues, offsetValues)])
    return binEdges, binMids

def set_priors(m):
    m.kern.lengthscales.prior = GPflow.priors.Gamma(0.1, 1.0)
    m.kern.variance.prior = GPflow.priors.Gamma(1.0, 1.0)
    m.kern.fixed = True
    
def plot_model(m, samples, gridResolution):
    intensities = []
    for s in samples:
        m.set_state(s)
        mu, _ = m.predict_y(m.X.value)
        intensities.append(mu)
    intensity = np.mean(intensities, 0)
    plt.figure(figsize=(6,6))
    plt.imshow(np.flipud(intensity.reshape(gridResolution, gridResolution).T),
               interpolation='nearest', extent=[0, 1, 0, 1])
    locs = load_pines()
    plt.plot(locs[:, 0], locs[:, 1], 'r.')

def build_model(gridResolution):
    binEdges, binMids = getGrid(gridResolution)
    counts = getCounts(gridResolution)
    kern = GPflow.kernels.Matern32(2)
    lik = GPflow.likelihoods.Poisson(binsize=binMids[1,1]-binMids[0,1])

    m = GPflow.gpmc.GPMC(binMids, counts, kern, lik)
    return m

def init_model(m):
    m.kern.lengthscales = 0.1
    m.kern.fixed = True
    m.optimize(maxiter=100)
    m.kern.fixed = False
```

```python
def changeGridsize(gridsize):
    print('grid is {} by {}'.format(gridsize,gridsize))
    m = build_model(gridsize)
    set_priors(m)
    init_model(m)
    burn = m.sample(200, epsilon=0.05, Lmin=8, Lmax=10)
    m.set_state(burn[-1])
    samples = m.sample(1000, epsilon=0.1, Lmin=8, Lmax=10, verbose=1)
    df = m.get_samples_df(samples)
    return m, samples, gridsize
```

```python
m, samples, gridsize = changeGridsize(16)
```

```python
plot_model(m, samples, gridsize)
```

## Let's use a larger binsize (reduce size of grid)

```python
m, samples, gridsize = changeGridsize(8)
plot_model(m, samples, gridsize)
```
